# What is Pandas?
Pandas is a Python library used for data manipulation, analysis, and cleaning.

1)Data Manipulation:

Changing, modifying, or restructuring data to make it usable (like adding/removing columns, merging data, etc.).

2)Data Analysis:

Examining data to find patterns, trends, or insights (like calculating averages, grouping data, etc.).

# What is Data Analysis?
Data Analysis means studying data to:

Understand what is happening

Find patterns and trends

Make decisions based on facts

# What do you mean by Patterns and Trends in Data?

1....Patterns:

Repeated behaviors or events found in data.

They show how data behaves regularly.

Example:

People buy more cold drinks in summer every year.

A customer always shops on weekends.

2...Trends:

A general direction of change in data over time.

Shows if something is increasing, decreasing, or stable.

Example:

Sales are increasing every month (upward trend).

Website traffic is dropping slowly (downward trend).






# Main Steps of Data Analysis:

| Step                      | Description                   | Example                       |
| ------------------------- | ----------------------------- | ----------------------------- |
| **1. Summarize**          | Get overall information       | `mean()`, `sum()`, `count()`  |
| **2. Group Data**         | Split data into categories    | `groupby()`                   |
| **3. Visualize**          | Use graphs or charts          | `plot()`, `hist()`, `bar()`   |
| **4. Find Relationships** | Check how things are related  | Correlation, `df.corr()`      |
| **5. Draw Conclusions**   | Use results to make decisions | Business reports, predictions |

3)Data Cleaning:

Fixing or removing incorrect, missing, or duplicate data to improve data quality.

# Why Pandas?
Handle tabular data (like Excel or CSV files).

Provides DataFrame and Series structures for data analysis.

Supports CSV, Excel, SQL, JSON, and other formats.

Allows filtering, sorting, grouping, merging, and visualization of data.

1️⃣ Forward Fill (method='ffill')
Meaning:
Copies the previous non-missing value into the next missing cell.

| Day | Temperature |
| --- | ----------- |
| Mon | 30          |
| Tue | **NaN**     |
| Wed | **NaN**     |
| Thu | 28          |

Result:

| Day | Temperature |
| --- | ----------- |
| Mon | 30          |
| Tue | 30          |
| Wed | 30          |
| Thu | 28          |



2️⃣ Backward Fill (method='bfill')
Meaning:
Copies the next non-missing value backward into the previous missing cell.

| Day | Temperature |
| --- | ----------- |
| Mon | 30          |
| Tue | **NaN**     |
| Wed | **NaN**     |
| Thu | 28          |

Result:

| Day | Temperature |
| --- | ----------- |
| Mon | 30          |
| Tue | 28          |
| Wed | 28          |
| Thu | 28          |

# 1️⃣2️⃣ Merging & Joining

| Function   | Purpose                                |
| ---------- | -------------------------------------- |
| `merge()`  | Merge tables (like SQL JOIN)           |
| `concat()` | Combine tables vertically/horizontally |


```python
# Creating two new dataframes for merging..

df01 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Ali', 'Sara', 'Zaid', 'Aman'],
    'Department': ['HR', 'IT', 'Finance', 'IT']
})
```

```python
df02 = pd.DataFrame({
    'ID': [3, 4, 5, 6],
    'Salary': [50000, 60000, 55000, 65000]
})
```

# Joins in Short...

| Join Type      | Code Example                               | What it does                                |
| -------------- | ------------------------------------------ | ------------------------------------------- |
| **Inner Join** | `pd.merge(df1, df2, on='ID', how='inner')` | Keeps **only matching IDs**                 |
| **Left Join**  | `pd.merge(df1, df2, on='ID', how='left')`  | Keeps **all from df1**, adds match from df2 |
| **Right Join** | `pd.merge(df1, df2, on='ID', how='right')` | Keeps **all from df2**, adds match from df1 |
| **Outer Join** | `pd.merge(df1, df2, on='ID', how='outer')` | Keeps **all rows from both**, fills NaNs    |


```python
# inner join

merged1 = pd.merge(df01, df02, on='ID', how='inner')

merged1
```

```python
# left join
merged2 = pd.merge(df01, df02, on='ID', how='left')
merged2
```

```python
# right join

merged3 = pd.merge(df01, df02, on='ID', how='right')
merged3
```

```python
# outer join

merged4 = pd.merge(df01, df02, on='ID', how='outer')
merged4
```

```python
merged4['Salary'] = merged4['Salary'].fillna(merged4['Salary'].mean())
```

```python
merged4
```

```python
merged4 = merged4.dropna(subset=['Name'])
```
