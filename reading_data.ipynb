What is a JSON file?

JSON (JavaScript Object Notation) is a data format used to store and exchange data between systems.

It is:

Lightweight and easy to read (for humans and computers)

Structured using key-value pairs (like Python dictionaries)

Commonly used for APIs, configurations, and data storage.

```python
# Example : 02

url1 = "https://en.wikipedia.org/wiki/Mobile_country_code"
```

```python
df1 = pd.read_html(url1,match='Country',header=0)
```

Pandas will only extract the table(s) that contain the word 'Country' in their text (either in the header or data).

If there are multiple tables in the HTML page, match helps you pick the correct one.

| Parameter  | Purpose                                   | Example Use       |
| ---------- | ----------------------------------------- | ----------------- |
| `match`    | Selects only tables **containing a word** | `match='Country'` |
| `header=0` | Uses **first row as column names**        | `header=0`        |

# Why is header important?

HTML tables sometimes have extra header rows (like titles or notes).

Specifying header=0 tells pandas to:

Skip any titles

Use the first row of the actual table as column names

# Creating a pickle file.

# What is Pickle?

Pickle is like saving your Python objects into a file so you can use them later.

It converts Python objects into a special format (binary) and saves them to a file.

Later, you can load them back exactly as they were.

# Why use Pickle?

Imagine you have a DataFrame, list, dictionary, or machine learning model.

You don't want to create it again and again.

So, you save it using Pickle.

Later, you can load it back instantly, even after closing your program.

# What is a Python Object?
In simple words:

A Python object is anything you create or use in Python.

| Example                | Python Object |
| ---------------------- | ------------- |
| `list`                 |  Object      |
| `dict`                 |  Object      |
| `DataFrame` (pandas)   |  Object      |
| `int`, `float`, `str`  |  Objects     |
| A trained **ML model** |  Object      |
| A **function**         |  Object      |
| A **class instance**   |  Object      |
